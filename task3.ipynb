```python
sn.pairplot(df,hue='Region')
```

```python
plt.figure(figsize=(7,5))
sn.countplot(x=df['Ship Mode'])
```

```python
plt.pie(list(df['Segment'].value_counts()),labels = list(df['Segment'].value_counts().keys()), radius =2,autopct = '%0.2f%%',colors=['red','green','blue'])
plt.show()
```

```python
#Plot for sales of each category (region-wise)
plt.figure(figsize=(10,5))
plt.title('Categories VS Region')
sn.countplot(x=df['Category'],hue=df['Region'])
plt.xticks()
```

```python
#Plot for number of sales of each product in sub-category.
plt.figure(figsize=(10,5))
sn.countplot(x=df['Sub-Category'],data=df)
plt.title('SUB-CATEGORY')
plt.xticks(rotation=90)
```

```python
#Pie-Chart for "SALES","DISCOUNT","PROFIT" for each "SEGMENT".  
shipmode1=df.groupby(['Segment'])['Sales','Discount','Profit'].sum()
shipmode1.plot.pie(autopct='%1.lf%%', label=shipmode1.index,subplots=True,figsize=(21,10))
plt.show()

```

```python
#Pie-Chart for "SALES","DISCOUNT","PROFIT" for each "CATEGORY".
smode1=df.groupby(['Category'])['Sales','Discount','Profit'].sum()
smode1.plot.pie(autopct='%1.lf%%', label=smode1.index,subplots=True,figsize=(21,10))
plt.show()

```

```python
#Pie-Chart for "SALES","DISCOUNT","PROFIT" for each "Region".
smode1=df.groupby(['Region'])['Sales','Discount','Profit'].sum()
smode1.plot.pie(autopct='%1.lf%%', label=smode1.index,subplots=True,figsize=(21,10))
plt.show()
```

```python
#LinePlot for Sales vs Profit of each Category.
plt.figure(figsize=(10,5))
sn.lineplot(x='Sales',y='Profit',data=df,hue='Category')
plt.title("Sales vs Profit")
```

```python
sales_profit=df.groupby(['Sub-Category'])['Profit','Sales'].agg(['sum'])
sales_profit.plot.bar(rot=0,figsize=(15,10))
plt.title('Sales Profit per Category',fontsize=35)
plt.xticks(rotation=90)
plt.show(block=True)
```
